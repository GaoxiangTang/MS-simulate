使用sympy自带的分段函数功能，计算A，M矩阵
其中
$$
A_{ml}=\int_0^T dt \int_0^t f_l(t_1)e^{i\delta_m t_1}dt_1\\
M_{ij}=-\frac{1}{2}\sum_{m}\eta_{m,j_1}\eta_{m,j_2}\int^T_0dt_1\int_0^{t_1}dt_2(f_i(t_1)f_j(t_2)+f_i(t_2)f_j(t_1))sin(\delta_m(t_2-t_1))
$$

```python

def f_l(tt, ll):
    return Piecewise((1, ((tt>tseg*ll)&(tt<=tseg*(ll+1)))), (0, True))

# calculate for M
Omega=1e9
for i in range(9):
    for j in range(9):
        t1, t2=symbols("t_1, t_2", positive=True)
        T=3e-4
        N=9
        delta=1e5
        tseg=T/N
        expr=Omega*(f_l(t1, i)*f_l(t2,j)+f_l(t2, i)*f_l(t1, j))*sin(delta*(t2-t1))
        intermediate=integrate(expr, (t2, 0, t1))
        print(i, j)
        print(integrate(intermediate, (t1, 0, T)))
        # sympy 很多时候会输出0的错误结果
        if i==j:
            print(Omega*(2*np.sin(delta*tseg)/delta**2-2*tseg/delta))
        else:
            print(Omega*(2*(np.cos(delta*tseg)-1)*np.sin(delta*tseg*abs(i-j)))/delta**2)
        print()
```
